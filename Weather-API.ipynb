```python
# getting 5-day, 3-hour forcast data from (https://rapidapi.com/community/api/open-weather-map/)
# better field descriptions: https://www.weatherbit.io/api/weather-forecast-5-day
# This code is coppied from Rapid API website

import requests
import pandas as pd
from datetime import date
from datetime import timedelta

def request_weather_data(city, country, API_key, url, headers):
	querystring = {"q":f"{city},{country}"}
	response = requests.request("GET", url, headers=headers, params=querystring)
	forecast_api = response.json()['list']
	weather_data = pd.DataFrame(extract_weather_info(forecast_api))
	return(weather_data)


def extract_weather_info(forecast_api):
    weather_info = []
    for forecast_3h in forecast_api: 
        weather_hour = {}
        # datetime utc
        weather_hour['datetime'] = forecast_3h['dt_txt']
        # temperature 
        weather_hour['temperature'] = forecast_3h['main']['temp']
        # wind
        weather_hour['wind'] = forecast_3h['wind']['speed']
        # probability precipitation 
        try: weather_hour['prob_perc'] = float(forecast_3h['pop'])
        except: weather_hour['prob_perc'] = 0
        # rain
        try: weather_hour['rain_qty'] = float(forecast_3h['rain']['3h'])
        except: weather_hour['rain_qty'] = 0
        # wind 
        try: weather_hour['snow'] = float(forecast_3h['snow']['3h'])
        except: weather_hour['snow'] = 0
        weather_hour['municipality_iso_country'] = city + ',' + country
        weather_info.append(weather_hour)
    return(weather_info)

city= "Berlin"
country = "DE"
API_key = "my-API-key"
url = "https://community-open-weather-map.p.rapidapi.com/forecast"
headers = {
	"X-RapidAPI-Host": "community-open-weather-map.p.rapidapi.com",
	"X-RapidAPI-Key": API_key}

weather_forcast = request_weather_data(city, country, API_key, url, headers)
weather_forcast['datetime'] = pd.to_datetime(weather_forcast['datetime'])
```
